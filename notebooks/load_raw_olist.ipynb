```python
from pathlib import Path
import duckdb

PROJECT_ROOT = Path("/home/mateus/dadosfera/MATEUS_WILAMIL_DDFTEC_012026")
RAW_PATH = PROJECT_ROOT / "data" / "raw"
DUCKDB_PATH = Path("/mnt/c/dadosfera/duckdb/olist.duckdb")

PROJECT_ROOT, RAW_PATH, DUCKDB_PATH
```

```python
RAW_PATH.exists(), DUCKDB_PATH.parent.exists()

DUCKDB_PATH.parent.mkdir(parents=True, exist_ok=True)
```

```python
con = duckdb.connect(DUCKDB_PATH)

con.execute("select 'conectado'").fetchall()
```

```python
con.execute("create schema if not exists raw;")
```

```python
def create_raw_table(table_name: str, file_name: str):
    file_path = RAW_PATH / file_name
    assert file_path.exists(), f"Arquivo não encontrado: {file_path}"

    con.execute(f"drop view if exists raw.{table_name}")
    con.execute(f"drop table if exists raw.{table_name}")

    con.execute(f"""
        create table raw.{table_name} as
        select *
        from read_csv_auto('{file_path}');
    """)

    print(f"✔ raw.{table_name} criado com sucesso")
```

```python
create_raw_table("olist_customers", "olist_customers.csv")
create_raw_table("olist_geolocation", "olist_geolocation.csv")
create_raw_table("olist_orders", "olist_orders.csv")
create_raw_table("olist_order_items", "olist_order_items.csv")
create_raw_table("olist_order_payments", "olist_order_payments.csv")
create_raw_table("olist_products", "olist_products.csv")
create_raw_table("product_descriptions", "product_descriptions.csv")
```
